```python
from typing import Dict, List, Optional
import requests



def get_speaker_names() -> List[str]:
    """スピーカーの名前のリストを取得する"""
    data = _get_speaker_data()
    return [speaker["name"] for speaker in data]


def get_speaker_name_by_id(speaker_id: int) -> Optional[str]:
    """スピーカーのIDから名前を取得する"""
    data = _get_speaker_data()
    for speaker in data:
        for style in speaker.get("styles", []):
            if style.get("id") == speaker_id:
                return speaker["name"]
    return None


def get_normal_speaker_names() -> Dict[int, str]:
    """ノーマルスピーカーのIDと名前の辞書を取得する"""
    data = _get_speaker_data()
    normal_speakers = {
        style.get("id"): speaker.get("name")
        for speaker in data
        for style in speaker.get("styles", [])
        if style.get("name") == "ノーマル"
    }
    return normal_speakers


def get_speaker_name_by_normal_id(speaker_id: int) -> Optional[str]:
    """ノーマルスピーカーのIDから名前を取得する"""
    normal_speaker_names = get_normal_speaker_names()
    name = normal_speaker_names.get(speaker_id)
    if name is not None:
        return name

    max_speaker_id = max(normal_speaker_names.keys())
    while speaker_id <= max_speaker_id:
        speaker_id += 1
        name = normal_speaker_names.get(speaker_id)
        if name is not None:
            return name
    return None

def _get_speaker_data() -> List[Dict]:
    """スピーカーデータを取得する"""
    data = _get_speaker_data.cache
    if data is None:
        host = "localhost"
        port = 50021
        response = requests.get(f"http://{host}:{port}/speakers")
        response.raise_for_status()
        data = response.json()
        _get_speaker_data.cache = data
    return data

def get_normal_speakers(speaker_id=4):
    speakers = _get_speaker_data()
    # ノーマルのキャラと番号の一覧を作成
    normal_speakers = {}
    normal_speakers = {
        style.get("id"): speaker.get("name")
        for speaker in speakers
        for style in speaker.get("styles", [])
        if style.get("name") == "ノーマル"
    }
    print(normal_speakers)

    while True:
        try:
            name = normal_speakers[speaker_id]
            # logger.info(f"{name}_{speaker_id}")
            return speaker_id
        except KeyError:
            speaker_id += 1
            # logger.debug("存在しないキャラクターIDです。次のキャラクターに移動します。")
        if speaker_id > max(normal_speakers.keys()):
            # logger.warning("全てのキャラクターを検索しました。最初のキャラクターに戻ります。")
            speaker_id = 1

def call_speaker_name_by_id(speaker_id=3):
    speakers = _get_speaker_data()
    normal_speakers = {
        style.get("id"): speaker.get("name")
        for speaker in speakers
        for style in speaker.get("styles", [])
        if style.get("name") == "ノーマル"
    }

    while True:
        try:
            if speaker_id in normal_speakers:
                # logger.info(f"ノーマルスピーカーのID {speaker_id} が見つかりました。")
                return speaker_id
            else:
                speaker_id += 1
                # logger.debug(f"{speaker_id} はノーマルスピーカーのIDではありません。次のIDに移動します。")
        except KeyError:
            # logger.warning("全てのキャラクターを検索しました。最初のキャラクターに戻ります。")
            speaker_id = 1


# キャッシュを持つ関数を定義
_get_speaker_data.cache = None

print(get_speaker_names())
print(get_speaker_name_by_id(15))
print(get_normal_speaker_names())
print(get_speaker_name_by_normal_id(15))
print(_get_speaker_data())
print(call_speaker_name_by_id(17))

# import timeit
# ll =  timeit.Timer(lambda: get_speaker_name_by_normal_id(10))
# gl =  timeit.Timer(lambda: get_speaker_name_by_id(8))

# print(timeit.timeit(get_speaker_names, number=1))
# print(ll.timeit(number=1))
# print(timeit.timeit(get_normal_speaker_names, number=1))
# print(gl.timeit(number=1))
# print(timeit.timeit(_get_speaker_data, number=1))
```
